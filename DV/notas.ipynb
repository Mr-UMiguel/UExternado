## Definición de derivado

Un derivado es un activo financiero cuyo valor depende del precio de un activo subyacente, se expresa como función del activo subyacente así

$$ V_{t} = f(S_{t}) $$ 

siendo $V_{t}$ el valor del derivado en el tiempo $t$ y $f(S_{y})$ una función que depende del precio $S$ del activo subyacente en el momento $t$, con $t \in \mathbb{N} $

## Tipos de derivados

 + Futuro: 
 
 Un futuro es un contrato obligante (CO) para la negociación del activo subyacente en una fecha futura $T$ y a un valor establecido $S$
 
 
 + Forward:
 
 Similar al futuro pero a diferencia de este, el forward se comercia en el mercado no OTC
 
 
 + Swaps:
 
 Un swap es un contrato obligante para el intercambio de flujos de capital, generalmente se presentan $\textbf{Interest Rate Swap}$ (IRS) y $\textbf{Credit Default Swap}$ (CDS)
 
 
 + Opciones:
 
 Una opción es una contrato NO obligante que otorga a su poseedor el derecho más no la obligación de negociar el subyacente en o antes de una fecha futura $T$ y por un valor determinado $K$ conocido como precio de mercado o strike price
 
 

>nota 1: En las opciones el poseedor del derecho paga una prima por adquirir este derecho

> nota 2:
+ largo $\rightarrow$ poseedor del derecho
+ corto $\rightarrow$ obligación de cumplir el derecho

> nota 3: 
 + derecho a comprar $\rightarrow$ call
 + derecho a vender $\rightarrow$ put

> nota 4:
 + Europeas $\rightarrow$ solo se pueden ejercer en T
 + Americanas $\rightarrow$ se pueden ejercer en $[0,T]$

## Uso de los derivados

 1. Cobertura:
 
    Se usan para cubrir diferentes tipos de riesgo, por ejemplo si una empresa local necesita adquirir 10.000 dólares incurre en riesgo de tipo de cambio
    
 2. Especulación:
 
    Un agente considera que un activo que un activo va a subir de precio los próximos 6 meses, por ende espera obtener ganancias 

```python
S = np.arange(0,100,1)
S0 = 50
profit = [price for price in S if price > S0]
loss = [price for price in S if price <= S0]

fig = plt.figure(figsize=(10,7))
plt.xlim((0,100))
plt.ylim((0,100))
plt.axhline(50, color="darkgrey")
plt.axvline(50, color="darkgrey", linestyle = '--')

#profit area
plt.plot(S[len(loss):],profit, color="green")
plt.fill_between(S[len(loss):],profit,50, color="green", alpha=0.2 )
plt.annotate("Profit area", xy=(80,65),xycoords='data',
            xytext=(80, 65), textcoords='data',color="green")


#loss area
plt.plot(S[:len(loss)],loss, color="salmon")
plt.fill_between(S[:len(loss)],loss,color="salmon", alpha=0.2 )
plt.annotate("Loss area", xy=(30,15),xycoords='data',
            xytext=(30, 15), textcoords='data',color="red")

plt.xlabel('S')
plt.ylabel('Balance')
plt.show()
```
