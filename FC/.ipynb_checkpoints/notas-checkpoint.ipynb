```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

from sklearn import preprocessing
from sklearn.cluster import KMeans

sns.set_theme(font="Times New Roman",font_scale=1)
```

```python
def load_data(symbols,start,end,plot):
    # Carga de datos

    data = pd.DataFrame({})
    for symbol in symbols:
        data_temp = yf.download(symbol, start=start, end=end,progress=False)
        data_temp = data_temp[['Adj Close']]
        data_temp = np.log(data_temp['Adj Close']) - np.log(data_temp['Adj Close'].shift(1))
        data_temp = data_temp.dropna()
        
        data = pd.concat([data,data_temp], axis=1)
        
    data.columns = symbols
    
    if plot==True:
        data.plot(figsize=(10,7))
        if len(symbols) > 3:
            plt.legend(loc='center right', bbox_to_anchor=(1,1),
                        fancybox=True, shadow=True, ncol=5)
        plt.title("logarithmic return", loc="left", fontdict = {'fontsize':18})
        plt.show()
    return data
```

## Funciones de densidad de probabilidad 

Una funcin de densidad de probabilidad se define como $f_{x}(x)$

Para calcular la probabilidad de que $X$ se encuentre entre un intervalo $ [a,b] $ $\mathbb{P}[a\leq X\leq b]$, calculamos la integral de $F_{x}(x)$ en ese intervalo

$$
\mathbb{P}[a\leq X\leq b] = \int_{a}^{b}f_{x}(x) dx
$$

```python
normal = np.random.normal(0,1,10000)
figure = plt.figure(figsize=(10,7))
sns.displot(normal, kind="kde")
plt.show()
```

## Funciones de distribucin acumulada de probabilidad 
Un funcin de distribucin acumulada de probabilidad se define como $F_{x}(x)$, dnde $ F_{x}(x):= \mathbb{P} [X\leq x] $
$$
\mathbb{P} [X\leq x] = \int_{-\infty}^{x}f_{x}(x)dx
$$

$$
f_{x}(x) = \frac{dF_{x}(x)}{dx}
$$

```python
sns.displot(normal, kind="ecdf")
plt.show()
```

## Distribuciones multivariadas

Dadas dos V.A continuas $X$ y $Y$ se define como la "distribucin conjunta" de estas variables

\begin{align}
f_{X,Y} (x,y) &:= \mathbb{P}[X \leq x, Y \leq y] \\
          & = \int_{-\infty}^{x}\int_{-\infty}^{y} f_{x,y}(x,y)dydx
\end{align}


```python
# si f_xy = x + y; 0 <= x <= 1; 0<= y <= 1
# %matplotlib inline

fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')
x = np.linspace(0,1,100)
y = np.linspace(0,1,100)
X, Y = np.meshgrid(x, y)
Z = X+Y
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('Distribuciones mltivariadas');
plt.show()
```

## Densidades marginales

para hallar la densidad de $X$ y $Y$ de la funcin de densidad conjunta $f_{X,Y}(X,y)$ debemos integrarla con respecto a cada V.A, por ejemplo, para hallar la densidad de $X$ se haya integrando la conjunta con respecto a Y

\begin{align}
f_{X}(x) &= \int_{-\infty}^{\infty} f_{X,Y}(X,Y) dy\\
f_{Y}(y) &= \int_{-\infty}^{\infty} f_{X,Y}(X,Y) dx\\
\end{align}

Dnde $0 \leq x \leq 1$ & $0 \leq y \leq 1$

## Distribuciones marginales

Dadas dos V.A $X$ y $Y$ continuas se define la densidad condicional de $X$ dado $Y$ como:

\begin{align}
f_{X|Y} &:= \frac{f_{X,Y}(xy)}{f_{Y}(y)} = \frac{x+y}{\frac{1}{2}+y}\\
f_{Y|X} &:= \frac{f_{X,Y}(xy)}{f_{X}(x)} = \frac{x+y}{\frac{1}{2}+x}
\end{align}

## Independencia de variables aleatorias 

Dos V.A $X$ y $Y$ son independientes sii:

\begin{align}
f_{X,Y}(x,y) &:= f_{x}(x)f_{y}(y)\\
E[X,Y] & = E[X]E[Y]\\
       & = \mu_{X} \mu_{Y}
\end{align}

# Modelos de volatilidad

## Momentos de una variable Aleatoria

Dado una variable aleatoria V.A $X$ con densidad $f_{X}(X)$ se define su valor esperado

\begin{align}
E[X] = \mu_{x} = \int_{-}^{} xf_{x}(x)dx
\end{align}

Tal que $X$ es una V.A continua; entonces:

El momento de orden $n$ con $n=1,2,3,...$ de $X$, sedefine como


\begin{align}
E[X^{n}] = \int_{-}^{} x^{n}f_{x}(x)dx
\end{align}

### El momento central de orden m de una V.A 
\begin{align}
E[(x-\mu_{x})^{n}] = \int_{-}^{} (x-\mu_{x})^{n}f_{x}(x)dx
\end{align}
 

### Momento de orden 2

Si $n=2$ tenemos la varianza de $X$.
\begin{align}
V[X] \equiv \sigma_{x}^{2} = \int_{-}^{} (x-\mu_{x})^{n}f_{x}(x)dx
\end{align}

Mide la dispersin, y definimos la desviacin estandar como

\begin{align}
\sigma_{x} = \sqrt{V[X]}
\end{align}

### Momento de orden 3
Si $n=3$, tenemos que se define el sesgo de x 
\begin{align}
S[X] = E[\frac{(x-\mu_{x})^{3}}{\sigma_{x}^{3}}]
\end{align}

Mide la simetra 

### Momento de orden 4
Si $n=4$, se define la curstsis de $X$

\begin{align}
K[X] = E[\frac{(x-\mu_{x})^{4}}{\sigma_{x}^{4}}]
\end{align}

mide el apuntamiento de $X$

## Estimadores de los momentos

### Estimador del valor esperado de una V.A
\begin{align}
E[X] = \hat{\mu_{x}}=\frac{1}{m}\sum_{i=1}^{m}x_{i}
\end{align}

### Estimador del momento de orden 2 (Varianza)
\begin{align}
\hat{\sigma_{X}^{2}}=\frac{1}{m-1}\sum_{i=1}^{m}(x_{i}-\hat{\mu_{x}})^{2}
\end{align}

### Estimador del momento de orden 3 (sesgo)
\begin{align}
\hat{S}_{x} = \frac{1}{m-1}\frac{1}{\hat{\sigma_{X}^{3}}}\sum_{i=1}^{m}(x_{i}-\hat{\mu_{x}})^{3}
\end{align}

### Estimador del momento de orden 4 (curtsis)
\begin{align}
\hat{S}_{x} = \frac{1}{m-1}\frac{1}{\hat{\sigma_{X}^{4}}}\sum_{i=1}^{m}(x_{i}-\hat{\mu_{x}})^{4}
\end{align}

## Hiptesis de Estacionariedad
Existe invarianza en el tiempo de los momentos estadsticos de la distribucin de una V.A


### Estacionariedad estricta
Un proceso aleatorio $\{X_{t}\}$ se dice "estrictamente estacionario" si para cualquier conjunto finito de tiempos $\{t_{1},t_{2},...,t_{k}\}$ y para cualquier intervalo $\tau$, se tiene que la distribucin conjunta del proceso $\{X_{t_{1}},X_{t_{2}}, ... , X_{t_{k}}\}$ es la misma que la distribucin conjunta del proceso $\{X_{t_{1+\tau}},X_{t_{2+\tau}}, ... , X_{t_{k+\tau}}\}$

### Estacionariedad dbil
Un proceso aleatorio $\{X_{t}\}$ se dice "dbilmente estacionario" o "estacionario en covarianza" si se cumple que:

  1. Tiene varianza finita $V[X_{t}]\leq \infty$
  2. Su media es constante $E[X_{t}]=\mu$



#### Covarianza
Mide la asociacin lineal entre variables

\begin{align}
cov(X,Y) = E[(X-\mu_{X})(Y-\mu_{Y})]
\end{align}

Propiedades de la covarianza

\begin{align}
cov(X,Y) &= E[(X-\mu_{x})(Y-\mu_{y})]\\
cov(\alpha X+\beta; \theta Y +\delta) &= \alpha \theta cov(X,Y)
\end{align}

```python
from IPython.core.pylabtools import figsize
X = np.random.normal(0,100,500)
Y = 3 + 2*X
Y1 = 3 - 2*X
Y2 = np.random.normal(0,100,500)

fig, axes = plt.subplots(2, 2, figsize=(10, 7))
fig.suptitle('Covarianza de X e Y')
sns.scatterplot(ax=axes[0,0],x=X,y=Y,color="cyan")
axes[0,0].set_title("covarianza > 0")
sns.scatterplot(ax=axes[1,0],x=X,y=Y1,color="salmon")
axes[1,0].set_title("covarianza < 0")
sns.scatterplot(ax=axes[0,1],x=X,y=Y2,color="skyblue")
axes[0,1].set_title("covarianza = 0")
plt.show()
```

#### Estimador de la covarianza 
\begin{align}
\hat{Cov(X,Y)} = \frac{1}{m-1}\sum_{i=1}^{m}(x_{i}-\hat{\mu_{x}})(y_{i}-\hat{\mu_{y}})
\end{align}

### Covarianza y coeficiente de correlacin

### Coeficiente de correlacin de Pearson 

El coeficiente de correlacin de Pearson $\rho$ se define como
\begin{align}
\rho &= \frac{cov(X,Y)}{\sigma_{X}\sigma_{Y}}
\end{align}

tal que $-1 \leq \rho(X,Y) \leq 1$


```python
corr = np.corrcoef(X,Y)
corr
```

### Autocovarianza 
Dado un proceso aleatorio $\{X_{t}\}$ y dos instantes de timepo $s$ y $t$ con $s< t$ se define su funcin de autocovarianza 

\begin{align}
\gamma_{X}(s,t) &:= Cov(X_{s},X_{t})\\
\gamma_{X}(h) &= cov(X_{t},X_{t+h})
\end{align}

$\gamma_{X}(h)$ se define como la autocovarianza con rezago $h$ tal que $h=0,+-1, +-2, ...$


#### Propiedades de la autocovarianza 
1. $\gamma_{X}(0) = cov(X_{t},X_{t}) = V(X_{t}) \geq 0$
2. $\gamma_{X}(h) = \gamma_{x}(-h) \quad \text{para todo h}$

# Clustering o agrupamiento


El objetivo central de las tcnicas de agrupamiento es identificar a partir de un conjunto de caractersitcas potenciales grupos.

## Distancia

Dado un conjunto de datos o caractersticas $\{x_{1}, x_{2}, ..., x_{n}\} \in \mathbb{R}^{n}$ una distancia (mtrica , norma) sobre este conjunto es una funcin $d:\mathbb{R}^{n}\times \mathbb{R}^{n}  \rightarrow \mathbb{R}$ y que cumple:

\begin{align*}
    d(x,y) &\geq 0 \quad \forall x,y\\
    d(x,y) &= 0 \quad sii \quad x=y\\
    d(x,y) &= d(y,x) \quad \forall x,y\\
    d(x,y) &\leq d(x,y) + d(z,y) \quad \forall x,y,z
\end{align*}

### Normas Lp:

La distancia entre dos puntos $||x-y||_{p} = (\sum_{i=1}^{n} |x_{i}-y_{i}|^{p})^{1/p} \quad ,p\geq1$

\begin{align*}
    p = 2 &\quad ||x-y||_{2} = (\sum_{i=1}^{n} |x_{i}-y_{i}|^{2})^{1/2} \quad &Euclidia \\
    p = 1 &\quad ||x-y||_{1} = \sum_{i=1}^{n} |x_{i}-y_{i}| \quad &Manhattan\\
\end{align*}

```python
table = pd.DataFrame({
    'ID': [1,2,3,4,5],
    'Score': [10,7,28,20,35]
})
table
```

### Construimos la matriz de proximidad

```python
#tomando la distancia euclidia
def similarity_matrix(table,distance):
    if distance == 'euclidian':
        euclidian = lambda p,q: np.sqrt((p-q)**2)
    n = table.shape[0]
    similarity_matrix = np.zeros((n,n))
    for row in range(table.shape[0]):
        for col in range(table.shape[0]):
            similarity_matrix[row,col] = euclidian(table.iloc[row,1],table.iloc[col,1])
    return similarity_matrix

similarity_matrix(table=table,distance='euclidian')
```

# Mtodos de Monte Carlo

El fudamento principal consiste en la generacin de nmeros aleatorios que se modelan bajo los valores de una variable de inters. La teora se basa en la ley de los grandes nmeros.

Los mtodo de montecarlo surgen como mtodos de integracin numrica.

## "Algunos Teoremas"

### La ley dbil de los grandes nmeros:

Si tenemos $N$ variables aleatorias $\{x_{1},x_{2},...,x_{n}\}$ i.i.d, y sea $\bar{X}_{n}$ el promedio de las variables aleatorias, entonces s $N$ es grande el promedio converge al valor esperado $E[X]$ de esas variables. 

\begin{align*}
    \mathbb{P}[\lim_{N \to \infty}(\bar{X}_{n}) = E[X]] = 1
\end{align*}

### Ley de Chebychev

Si $X$ es v.a con valor esperado $\mu$ y $V[X]=\sigma^{2}$ Entonces

\begin{align*}
    \mathbb{P}[|x-\mu|>x]\leq\frac{\sigma^{2}}{x}
\end{align*}

### Teorema fundamental de Monte Carlo

Si $\bar{X}_{n}$ se calcula con $N \to \infty$ entonces

\begin{align*}
    E[\bar{X}_{n}] &= E[X]\\
    V[\bar{X}_{n}] &= \frac{V[X]}{N}
\end{align*}

## Mtodo de montecarlo para estimar el valor de $\pi$

```python
def pi_aprox(n):
    x, y = np.random.uniform(low=0, high=1, size=(2,n))
    int = (x**2+y**2) <= 1
    ext = (x**2+y**2) > 1
    pi =  4 * (sum(int)/n)
    plt.scatter(x[int],y[int],color='skyblue')
    plt.scatter(x[ext],y[ext],color='salmon')
    plt.axis('square')
    print(f'El valor estimado de Pi es: {pi}')
```

```python
pi_aprox(5000)
```

## Mtodos de montecarlo para integracin numrica 

Sea $X$ una v.a tal que $X \sim U[1,b]$

\begin{align*}
    f_{x}(x) &= \frac{1}{b-a} \quad \text{;} \quad a\leq x \leq b\\
    E[g(x)]  &= \int_{a}^{b} g(x)\frac{1}{b-a}dx\\
    E[g(x)]  &= \frac{1}{b-a} \int_{a}^{b}g(x)dx\\
    \int_{a}^{b}g(x)dx &= (b-a)E[g(x)]\\
             &\approx (b-a) \frac{1}{n}\sum_{i=1}^{n}g(x_{i})
\end{align*}



#### ejercicio, aproximar:

\begin{align*}
    \int_{0}^{1}xe^{x}dx &\approx (1-0)\frac{1}{n}\sum_{i=1}^{n}x_{i}e^{x_{i}} \quad ; \quad x_{i} \sim U[0,1]
\end{align*}

```python
func = lambda x: x*np.exp(x)
x =  np.random.uniform(0,1,size=10000)
Int = np.mean(func(x))
Int
```

# Simulacin de procesos estocsticos

Un proceso estocstico $\{X_{t}\}_{t \in I}$ es una coleccin de variables aleatorias indexadas en el  tiempo.

    Nota: Los proceso que ms se trabajan en finanzas son los procesos Binominales y los procesos del movimiento Browniano estndar o proceso de Wiener

Los P.E se utilizan para modelar el componente de aleatoriedad de una variable.

## Proceso Binomianles $\{M_{t}\}_{t=0,1,...,T}$
Un proceso se dice binomial si 


Donde $X_{i}$ son i.i.d tal que,
\begin{align*}
M_{o} &= 0\\
M_{t} &= \sum_{i=1}^{t}X_{i}\\
X_{i} &= 
     \begin{cases}
                    u \to p\\
                    d \to 1-p
    \end{cases} \quad ; \quad p \in [0,1]\\

E[M_{t}] &= t[up + d(1-p)]\\
V[M_{t}] &= \sum_{i=1}^{t}V[X_{i}] + 0 \quad ; \quad X_{i} \sim iid\\
M_{t} &\sim Binomial \\
M_{t+1} &= M_{t} + X_{t+1}
\end{align*}



```python
# Definimos el valor de u, d y p
u = 2
d = -2
p = 0.3
t = 500

random = np.random.uniform(0,1,t)
Xi = [u if i <= p else d for i in random]
Mt = np.zeros(t)
for i in range(t-1):
    Mt[i+1] = Mt[i] + Xi[i]
pd.Series(Mt).plot()
```

```python
u = 3
d = -1
p = 0.3
t = 100
Nsim = 30
M = np.zeros((t+1,Nsim))
for j in range(Nsim):
    for i in range(t):
        random = np.random.uniform(0,1)
        if random <= p:
            xi = u
        else:
            xi = d
        M[i+1,j] = M[i,j]+ xi
pd.DataFrame(M).plot(legend=False)
```

## Caminata aleatoria

Una caminata aleatoria simtrica es un proceso Binomial en el cul se tiene que:
\begin{align*}
X_{i} =  
    \begin{cases}
        1 \to 1/2\\
        -1 \to 1/2\\
    \end{cases}
\end{align*}

```python
x1 = 1
x2 = 0
x3 = -1
p = 0.25
t = 200
Nsim = 2
M = np.zeros((t+1,Nsim))
for j in range(Nsim):
    for i in range(t):
        random = np.random.uniform(0,1)
        if random <= p:
            xi = x1
        elif p <= random < 0.75:
            xi = x2
        else:
            xi = x3
        M[i+1,j] = M[i,j]+ xi
pd.DataFrame(M).plot(legend=False)
```

## Movimiento Browniano Estndar 

El Movimiento Browniano Estndar $\{W_{t}\}$ es un tipo de proceso que modela aleatoriedad en tiempo continuo 

Definicin:

Un proceso estocstico $\{w_{t}\}_{t\geq 0}$ se dice Movimiento Browniano Estndar, si cumple:

\begin{align*}
    i)  \quad & w_{0} = 0\\
    ii) \quad &\text{Las trayectorias son continuas}\\
    iii)\quad&\text{Dados dos instantes s y t con s < t, se tiene que}\\
    iv) \quad &\quad W_{t} - W_{s} \sim N(0,t-s)\\
    v)  \quad&\text{Dados dos instantes s y t con s < t, se tiene que} W_{t} - W_{s} \text{es independiente de} \mathcal{F}
\end{align*}

Propiedades

\begin{align*}
    1)&  \quad  W_{t} \sim N(0,t)\\
    2.1)& \quad W_{t + dt} - w_{t} \sim N(0,dt)\\
    2.2)& \quad \text{sea } Z \sim N(0,1) \rightarrow \sqrt{dt}Z \sim N(0, dt)
        & W_{t+dt}-W_{t} =^{d} \sqrt{dt}Z \rightarrow W_{t+dt} = W_{t} + \sqrt{dt}Z \rightarrow N(0,\sqrt{dt}Z)
\end{align*}


## Proceso asociados al Movimiento Browniano
### Movimiento Browniano con Tendencia 
\begin{align*}
    X_{t} &= \alpha t + \sigma W_{t} \quad ; \quad \alpha, \sigma > 0 \quad \text{constantes}
\end{align*}

Una de las aplicaciones en Finanzas es para modelar Flujos de Caja Descontados

Propiedades
\begin{align*}
    E[X_{t}] &= \alpha t\\
    V[X_{t}] &= \sigma^{2} t\\
    X_{t} & \sim N(\alpha t , \sigma^{2} t)\\
    X_{t+dt} &= \alpha (t+dt) + \sigma W_{t+dt}\\
    X_{t+dt}-X_{t} &= \alpha dt + \sigma (X_{t+dt}- X_{t})\\
    X_{t+dt} &= X_{t} + \alpha dt + \sigma \sqrt{dt}Z \quad (6.5)
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Parmetros iniciales
alpha, sigma, T, n = 1, 0.4, 1, 10000
dt = T/n
Nsim = 50

# Creamos la matriz de trayectorias
X = np.zeros((n+1,Nsim))
for j in range(Nsim):
    for i in range(n):
        Z = np.random.randn() # de (6.5)
        X[i+1,j] = X[i,j] + alpha*dt + sigma*np.sqrt(dt)*Z  
```

```python
#Graficamos
tiempo = np.linspace(0,T,n+1)
Esp_X = alpha*tiempo # Valor esperado
Des_Xt = sigma*np.sqrt(tiempo) # Desviacin estndar

#Intervalos de confianza
#(\bar{X} - z_{\alpha / 2}\sigma; \bar{X} + z_{\alpha / 2}\sigma) 

fig = plt.figure(figsize=(10,7))
for j in range(Nsim):
    plt.plot(tiempo,X[:,j])
    
plt.plot(tiempo,Esp_X, c='k')
plt.plot(tiempo,Esp_X+Des_Xt,c='k')
plt.plot(tiempo,Esp_X-Des_Xt,c='k')
plt.show()
```
